# 01 · Análisis Exploratorio de Datos (EDA) — AnemIA
Dataset real de anemia infantil en Puno. Explora distribución de clases, hemoglobina y altitud.

```python
import os, sys
sys.path.insert(0, os.path.abspath('..'))
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from ml.preprocessing_pipeline import load_dataset, CLASS_ORDER
df = load_dataset('../data/raw/dataset.csv')
print(df.shape)
df.head()
```

```python
# Distribución de clases (desbalanceada)
df['Dx_anemia'].value_counts().reindex(CLASS_ORDER).plot.bar(color='#0d9488')
plt.title('Distribución de diagnósticos'); plt.ylabel('Casos'); plt.tight_layout(); plt.show()
```

```python
# Relación altitud vs hemoglobina observada y corregida (Hbc)
fig, ax = plt.subplots(1, 2, figsize=(11,4))
sns.scatterplot(data=df.sample(3000), x='AlturaREN', y='Hemoglobina', hue='Dx_anemia', s=10, ax=ax[0])
sns.scatterplot(data=df.sample(3000), x='AlturaREN', y='Hbc', hue='Dx_anemia', s=10, ax=ax[1])
ax[0].set_title('Hb observada'); ax[1].set_title('Hbc corregida'); plt.tight_layout(); plt.show()
```
