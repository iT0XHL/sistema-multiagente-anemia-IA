```python
import os, sys
sys.path.insert(0, os.path.abspath('..'))
os.chdir('..')  # ejecutar desde la raíz del repo
from ml import train_random_forest
metrics = train_random_forest.train('data/raw/dataset.csv')
metrics
```

```python
import numpy as np, matplotlib.pyplot as plt
cm = np.array(metrics['confusion_matrix']); classes = metrics['classes']
plt.imshow(cm, cmap='Blues'); plt.colorbar()
plt.xticks(range(len(classes)), classes, rotation=45); plt.yticks(range(len(classes)), classes)
plt.title('Matriz de confusión · Random Forest'); plt.tight_layout(); plt.show()
```
